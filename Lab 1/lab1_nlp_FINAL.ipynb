# 1. How to Read Academic Papers?

We will have to read several papers in this class (15-20 - why?). Watch this lecture from Andrew Ng and summarize his strategy on how to read research papers.

https://www.youtube.com/watch?v=733m6qBH-jI

You may use ChatGPT (or another LLM of your choice) to help you summarize the main points of this lecture. 

* Compile a list of papers from typical sources
* Create a line plot starting at 0% to 100%
* create a goal of how much of each paper to read (such as 10%)
* Skip around and read each paper at the beginning (10% of the paper)
* This allows you to decide early which papers are duds
* Also allows you to see if the research paper is worth reading all the way through. Looks at citations and sees which papers are connected to each other.

When starting to read papers, read the titles, abstract, and the figures (saves time to assess what the paper is about)
Afterwards, go to the intro and conclusion and look more closely at the figures and skim whats relevant. Allows you to see if the author has convinced you to delve deeper into their reseach. Then read it again and skip through things you do not understand. 

# 2. Define the terms.

- Corpus - set of texts used for analysis, training, and validation. Datrabse for linguistic data.
- Document - Block of text that is considered a single unit of data. Examples are emails, articles, chapters. In a corpus, each entry is typically a document.
- Sentence - set of words that convey a complete thought. 
- Word - A discrete element, typically separated by spaces in text. The primary unit of meaning in NLP context.
- Stop words - Commonly used words that are typically ignored when indexing entries. Considered little value to process. Example words are "the", "a", "an", "in".
- Punctuation - Helps organize and structure the text. Provides a way to split sentences or different understanding of the text. Example punctuations are "." "," ";".
- Stemming - Simplify a word to its word stem. Examples are "fishing", "fished", "fishter" reduced to the stem "fish"
- Lemmatization - reduce words to normal dictionary definition, or its base lemma. Examples are "is", "are", "am" which all can mean "be"

```python
import nltk

from nltk.corpus import *
import string
```

```python
#define function to find averages
def avg(text):
    g_num_chars = 0
    g_num_words = 0
    g_num_sents = 0
    g_num_vocab = 0

    for fileid in text.fileids():
         g_num_chars += len(text.raw(fileid))
         g_num_words += len(text.words(fileid))
         g_num_sents += len(text.sents(fileid))
         g_num_vocab += len(set(w.lower() for w in text.words(fileid)))

    g_avg_word_length = round(g_num_chars/g_num_words)
    g_avg_sent_length = round(g_num_words/g_num_sents)
    g_avg_num_repeat = round(g_num_words/g_num_vocab)

    print("Average Word Length: " + str(g_avg_word_length), ", Average Sentence Length: " + str(g_avg_sent_length), ", Average Repeats: " + str(g_avg_num_repeat))
    #print("Words: " + str(num_words), ", Sentences: " + str(num_sents))
```

```python
stopword = stopwords.words('english')
print(stopword)
```

```python
punctuations = string.punctuation
print(punctuations)
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

```python
from nltk.stem import *
from nltk.stem.porter import *

stemmer = PorterStemmer()
```

```python
wnl = WordNetLemmatizer()
```

# 4. Statistics

Determine the value for the formula $\mathbb{E}_{(x,y) \sim (U(-1,1), N(0,1))} \left[ x y^2 \right]$, and using numpy, validate your answer.

### x ~ U(-1,1)
(b+a)/2 , U(a,b)
(1-1)/2 = 0
E(x) = 0;

### y ~ N(0,1)
N(mu,sigma^2)
Var(y) = E(y^2)-E(y)^2
E(y^2) = Var(y) + E(y)^2

Var(y) = 1; E(y)^2 = 0
E(y^2) = 1

**(xy^2) = 0*1 = 0**

```python
import numpy as np

num_samples = 100000

# samples for x ~ U(-1,1)
x_samples = np.random.uniform(-1, 1, num_samples)

# samples for y ~ N(0,1)
y_samples = np.random.normal(0, 1, num_samples)

products = x_samples * y_samples**2

# average to estimate expectation
estimated_expectation = np.mean(products)
estimated_expectation
```
